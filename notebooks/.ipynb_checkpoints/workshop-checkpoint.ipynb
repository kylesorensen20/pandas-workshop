###  The Data Structures of `pandas`

```python
# There are two primary data structures in pandas -
# the `DataFrame` and the `Series`
df = pd.DataFrame({'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
# note: the conceptual structure of a DataFrame 
# can be thought of as a dictionary or hashmap
# with string keys and list values
s = pd.Series([1, 2, 3])
print(s)
s = df.A
print(s)
# note: each column in a DataFrame is a labeled
# Series object, this is the link between the two 
# basic data structures in pandas
```

### Viewing Data and Metadata

```python
# There are three functions of interest here:
print(df.head(2)) # head(n) displays the FIRST n rows of data
                    # from our DataFrame object
print(df.tail(2)) # tail(n) displays the LAST n rows of data
                    # from our DataFrame object
print(df.info()) # displays metadata for each column in our
                    # DataFrame object (null info!!)
```

### Indexing, Selecting and Filtering by Row and Column

```python
## First, we show how to select by column...

# Selecting a column using . notation
A_series = df.A

# Selecting a column using subscription notation (should be more familiar)
# Selecting multiple columns, can be useful to "declutter" your data or 
# more directly compare columns
narrowed_df1 = df[["A", "B"]]

# Selecting columns in a range using .loc() and columns at certain indices using .iloc()
narrowed_df2loc = df.loc[:,'A':'C']
narrowed_df2iloc = df.iloc[:,[0,1,2]] # note that this achieves the same thing 
                                        # as the subscription notation above!
# note: can select columns using a regular expression (i.e. only
# columns starting with a certain letter) or by data type (i.e keep
# only the numeric columns)

## Next, we show how to select by row...

# Filtering rows for a specific Boolean condition
df_largeA = df[df.A > 2]

# Dropping duplicate observations from a DataFrame
df_norep = df.drop_duplicates()

# Randomly sample (select) (1-frac)100% of the rows in the DataFrame
df_frac = df.sample(frac=0.5)

# Randomly sample (select) n of the rows in the DataFrame
df_sample = df.sample(n=2)
# note: the most common operation here is filtering, but be careful using
# 'and' and 'or' on Series objects, have to use bitwise operators instead
```

### Merge, join, and concatenate

```python
# concat - "stacks" DataFrames end over end
df1 = pd.DataFrame(
     {
         "A": ["A0", "A1", "A2", "A3"],
         "B": ["B0", "B1", "B2", "B3"],
         "C": ["C0", "C1", "C2", "C3"],
         "D": ["D0", "D1", "D2", "D3"],
     })
df2 = pd.DataFrame(
     {
         "A": ["A4", "A5", "A6", "A7"],
         "B": ["B4", "B5", "B6", "B7"],
         "C": ["C4", "C5", "C6", "C7"],
         "D": ["D4", "D5", "D6", "D7"],
     })
df3 = pd.DataFrame(
     {
         "A": ["A8", "A9", "A10", "A11"],
         "B": ["B8", "B9", "B10", "B11"],
         "C": ["C8", "C9", "C10", "C11"],
         "D": ["D8", "D9", "D10", "D11"],
     })
concat_result = pd.concat([df1, df2, df3]).reset_index().drop(columns=["index"])

# merge (example is equivalent to an inner join)
# argument "how" has options inner, outer, cross, right and left
# argument "validate" can be used to ensure a "good" or one-to-one join
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
merge_result = pd.merge(left, right, on="key")

# join (example is equivalent to an inner join)
# argument structure is essentially the same as that
# for merge above, but syntax is perhaps more intuitive
# for people who are used to OOP vs. FP
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)
right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
join_result = left.join(right)
```

### Pivot Tables and Data Reshaping

```python
# Pivot tables and data reshaping
import pandas._testing as tm
def unpivot(frame):
    N, K = frame.shape
    data = {
        "value": frame.to_numpy().ravel("F"),
        "variable": np.asarray(frame.columns).repeat(N),
        "date": np.tile(np.asarray(frame.index), K),
    }
    return pd.DataFrame(data, columns=["date", "variable", "value"])
df = unpivot(tm.makeTimeDataFrame(3))
pivoted = df.pivot(index="date", columns="variable", values="value")
# note: arguments to pivot can be lists (i.e. multi-indexed pivot tables)
```

## "Advanced" Tasks for This Workshop
Using the skills we learned above along with some additional functionality from other libraries, we will complete the following tasks...
* Generate a pivot table displaying average home price w.r.t. the number of bathrooms and number of bedrooms and create a 3D visualization of the summarized data
* Construct a time series model for housing prices with a 6 month forecast

### Task 1: Pivot Table and 3D Plot of Average Home Prices

```python
# Construct "rounded bathrooms" variable to limit number of buckets
home_data["bathrooms_rd"] = np.round(home_data.bathrooms, decimals=0)
# Apply a group by statement with "mean" as the aggregate function
mean_cost_numrooms = home_data.groupby(["bedrooms", "bathrooms_rd"])["price"].mean().reset_index()
# Apply pivot command according to desired output
mean_cost_numrooms_pivot = mean_cost_numrooms.pivot(index="bedrooms",
                                                    columns="bathrooms_rd",
                                                    values="price")
# Create 3D scatter plot
fig = plt.figure()
ax = plt.axes(projection = '3d')
x = mean_cost_numrooms['bedrooms']
y = mean_cost_numrooms['bathrooms_rd']
z = mean_cost_numrooms['price']
ax.set_xlabel("Bedrooms")
ax.set_ylabel("Bathrooms (Rounded)")
ax.set_zlabel("Home Price")
ax.scatter(x, y, z)
ax.view_init(azim=20,elev=10)
plt.show()
```

### Task 2: Holt-Winters' Forecast for Average Home Prices

```python
# Create a "month" variable using time series features of pandas
# to "floor" the date variable
home_data["month"] = home_data["date"].dt.date + pd.tseries.offsets.MonthBegin(0)
# Get the average home price by month
home_price_monthly = home_data.groupby(['month'])['price'].mean().reset_index()
# Forecast using Holt Winters model (6 months forward) from statsmodels package
from statsmodels.tsa.holtwinters import ExponentialSmoothing
fitted_model = ExponentialSmoothing(home_price_monthly['price'],
                                    trend='mul',seasonal='mul',
                                    seasonal_periods=2).fit()
fc_prices = fitted_model.forecast(6)
fitted_prices = pd.concat([fitted_model.fittedvalues, fc_prices])
start_date = home_price_monthly.iloc[0].month
end_date = start_date + pd.offsets.DateOffset(months=18)
monthly_range = pd.date_range(start=start_date, end=end_date,
                             periods=19) + pd.tseries.offsets.MonthBegin(0)
fitted_prices = pd.DataFrame({'month': monthly_range.date, 'price': fitted_prices})
df.iloc[9].month = pd.to_datetime('2015-03-01')
# Create plot of forecast
plt.plot(home_price_monthly['month'], home_price_monthly['price'], label='Actual')
plt.plot(fitted_prices['month'], fitted_prices['price'], label='Forecast')
plt.legend()
plt.title('Holt Winters Forecast for Home Prices (Monthly)')
plt.xticks(rotation=45)
plt.show()
# note: pandas recently removed some support for monthly indexing (very sad), so this
# ended up taking a lot longer than expected but oh well
# note: this is probably not the optimal model choice (note the convergence warning)
# but that is beyond the scopeof this workshop and could probably constitute a whole project
```

### Task 3: Interpolated Heatmap of Average Home Prices

```python
## Task 3: Heatmap of Average Home Prices
# First get the geographic means of our dataset
geo_price_data = home_data[["lat","long","price"]].groupby(["lat","long"])['price'].mean().reset_index()
# Interpolate over the coordinate region
from scipy.interpolate import interp2d
x_list = np.array(geo_price_data["long"])
y_list = np.array(geo_price_data["lat"])
z_list = np.array(geo_price_data["price"])
f = interp2d(x_list, y_list, z_list, kind="linear")
x_coords = np.arange(min(x_list), max(x_list) + 1)
y_coords = np.arange(min(y_list), max(y_list) + 1)
c_i = f(x_coords, y_coords)
plt.imshow(
    c_i,
    extent=[min(x_list), max(x_list), min(y_list), max(y_list)],
    origin="lower",
    interpolation="bicubic", cmap='Reds'
)
# Produce interpolated heatmap
plt.scatter(x_list, y_list, 400, facecolors="none")
import matplotlib.cm as cm
plt.colorbar(cm.ScalarMappable(cmap='Reds'))
plt.show()
# Note: as before, this is likely NOT the optimal interpolation method (search "kriging"
# or "gaussian process modeling" for more optimal spatial models)
# Bonus: the library called geopandas would allow us to add geographic features
# over the coordinate region (pretty cool!), but for the record, R has a couple of
# libraries called "fields" and "maps" which are signficantly more user friendly!!
```
